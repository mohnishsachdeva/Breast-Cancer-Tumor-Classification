# Fit_Transform

### The (fit method) is calculating the mean and variance of each of the features present in our data.
### The (transform method) is transforming all the features using the respective mean and variance.

# transform()
### Using the transform method we can use the same mean and variance as it is calculated from our training data to transform our test data. Thus, the parameters learned by our model using the training data will help us to transform our test data.

# fit() 
### just calculates the parameters (e.g. μ and σ in case of StandardScaler) and saves them as an internal object's state. Afterwards, you can call its transform() method to apply the transformation to any particular set of examples.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
algorithms = [
                GaussianNB,
                LogisticRegression
             ]
```

```python
df_algorithms = pd.DataFrame(columns=["Model", "Train Accuracy", "Test Accuracy","TP","TN","FP","FN","Accuracy","Precision","Recall","F1-Score","Specificity"])

confusion_matrices = {}
```

```python
def autoML(algorithm): 

    model = algorithm().fit(X_train, y_train)
#     print('model', model)
    train_acc = model.score(X_train, y_train)
#     print('train_acc',train_acc)
    model_name = algorithm.__name__
#     print('model_name',model_name)
    
    y_pred = model.predict(X_test)
#     print('y_pred',y_pred)
    test_acc = accuracy_score(y_test, y_pred)
#     print('test_acc',y_pred)

    actual_class = np.array(y_test)
    actual_class = actual_class.reshape(actual_class.shape[0])
    
    confusion_matrix_df = pd.crosstab(y_pred,actual_class, colnames = ['actual'], rownames = ['predicted'], margins = True)
    
    tp = confusion_matrix_df[1][1]
    fn = confusion_matrix_df[1][0]
    fp = confusion_matrix_df[0][1]
    tn = confusion_matrix_df[0][0]
    Accuracy  = ((tp+tn)/(tp+tn+fp+fn))
    Precision = (tp/(tp+fp))
    Recall    = (tp/(tp+fn))
    F1_score  = (2*(((tp/(tp+fp))*(tp/(tp+fn)))/((tp/(tp+fp))+(tp/(tp+fn)))))
    Specificity = (tn/(tn+fp))
   
    return model_name, train_acc, test_acc, confusion_matrix_df, tp, tn, fp, fn, Accuracy, Precision, Recall, F1_score, Specificity
```

```python
for alg in algorithms:
    
    model_name, train_acc, test_acc, confusion_matrix, true_positive, true_negative, false_positive, false_negative,Accuracy, Precision, Recall, F1_score, Specificity = autoML(alg)
    
    df_algorithms = df_algorithms.append(
                                        { "Model" : model_name,
                                          "Train Accuracy": train_acc,
                                          "Test Accuracy": test_acc,
                                          "TP" : true_positive,
                                          "TN": true_negative,
                                          "FP" : false_positive,
                                          "FN" : false_negative,
                                          "Accuracy": Accuracy,
                                          "Precision" : Precision,
                                          "Recall": Recall,
                                          "F1-Score": F1_score,
                                          "Specificity": Specificity
                                         }, ignore_index=True)
    confusion_matrices[model_name] = confusion_matrix
```
